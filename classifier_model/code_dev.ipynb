```python
import re
import json
from tqdm import tqdm
```

```python
file = './data/attribute_test.data' # _stratified
output_file = './data/test.json'

# file = './data/attribute_train_stratified.solution'
# output_file = './data/train_sol.json'
```

```python
data = []
with open(file, 'r') as f:
        for line in tqdm(f):
            datapoint = {}
            line = line.strip().strip('{}')
            pattern = r'"([^"]+)":\s?(?:(\d+(?:\.\d+)?)|"([^"]+)")' # match key: value or key: "value"
            attrs = re.findall(pattern, line)
            for attr in attrs:
                key = attr[0].strip().replace('\"', '') # remove " from key
                if key == 'indoml_id':
                    value = int(attr[1]) if attr[1] != '' else attr[2].strip().replace('\"', '')
                else:
                    value = str(float(attr[1])) if attr[1] != '' else attr[2].strip().replace('\"', '')
                datapoint[key] = value
            data.append(datapoint)
            
with open(output_file, 'w') as f:
    json.dump(data, f, indent=4)
```
